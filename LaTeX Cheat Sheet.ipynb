## LaTeX Cheat Sheet in Jupyter Notebook

### Basic Syntax
- **Inline Equations:** Wrap with single dollar signs: `$...$`
  - Example: `$E=mc^2$` renders as $E=mc^2$
- **Display Equations:** Wrap with double dollar signs: `$$...$$`
  - Example: `$$E=mc^2$$` renders as $$E=mc^2$$

### Common Mathematical Notations

- **Subscripts and Superscripts**
  - Subscript: `$a_i$` renders as $a_i$
  - Superscript: `$a^i$` renders as $a^i$
  - Combined: `$a_i^j$` renders as $a_i^j$

- **Mathematical Functions**
  - Sqrt: `$\sqrt{x}$` renders as $\sqrt{x}$
  - nth root: `$\sqrt[n]{x}$` renders as $\sqrt[n]{x}$
  - Trig functions: `$\sin, \cos, \tan$` render as $\sin, \cos, \tan$

- **Fractions and Binomials**
  - Fraction: `$\frac{a}{b}$` renders as $\frac{a}{b}$
  - Binomial: `$\binom{n}{k}$` renders as $\binom{n}{k}$

- **Summations and Integrals**
  - Summation: `$\sum_{i=1}^{n} a_i$` renders as $\sum_{i=1}^{n} a_i$
  - Integral: `$\int_{a}^{b} x^2 dx$` renders as $\int_{a}^{b} x^2 dx$

- **Special Symbols**
  - Infinity: `$\infty$` renders as $\infty$
  - Partial derivative: `$\partial$` renders as $\partial$
  - Element of: `$\in$` renders as $\in$
  - Not equal: `$\neq$` renders as $\neq$
